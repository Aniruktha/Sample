```python
from re import S
from collections import deque
import json

class Machine:
  def __init__(self,id,s_id,init_para,cooldown,flac,n):
    self.m_id=id
    self.step_id=s_id
    self.ideal_time=cooldown
    self.ini_parameters=init_para
    self.curr_parameters=init_para
    self.end_time=0
    self.processed_wafers=0
    self.n=n
    self.fluctuation=flac


class Wafer:
  def __init__(self,id,step,time):
    self.w_id=id
    self.step_id=step
    self.ready_time=0
    self.processing_time=time
    self.completed_step=[]


def dependancy(relations):
  dag={}
  for i in relations:
    if i["dependency"]:
      dag[i["id"]]=i["dependency"]
  return dag

def get_step(steps,dag):
  step=steps[0]
  return step
  # if step not in dag.keys():
  #   return step
  # for s in dag[step]:
  #   get_step(s,dag)


def inRange(parameters,curr_para):
  c=0
  for key in parameters:
    curr=curr_para[key]
    actual=parameters[key]
    if curr<=actual[1] and curr>=actual[0]:
      c+=1
  if c==len(parameters):
    return True
  return False


def cooldown(m):
  m.end_time+=m.ideal_time
  m.curr_parameters=m.ini_parameters


def allPossibleSteps(w,dag,all_steps):
  for i in w.step_id:
    if i not in dag.keys():
      all_steps.add(i)
    else:
      deps=dag[i]
      if all(d in w.completed_step for d in deps):
        all_steps.add(i)


with open("/content/Milestone1.json","r") as f:
  content=json.load(f)


steps_parameters={}
machines={}
wafers=[]
current_time=0
event_queue=deque()

res=[]

#initialising
for key in content:
  if key=="steps":
    dag=dependancy(content[key])
    for i in content[key]:
      steps_parameters[i["id"]]=i["parameters"]

  if key=="machines":
      for m in content[key]:
        obj=Machine(m["machine_id"],m["step_id"],m["initial_parameters"],m["cooldown_time"],m["fluctuation"],m["n"])
        if m["step_id"] not in machines:
          machines[m["step_id"]]=[]
        machines[m["step_id"]].append(obj)

  if key=="wafers":
    type="W1"
    c="0"
    for w in content[key]:
      for i in range(w["quantity"]):
        if type==w["type"]:
          c=str(int(c)+1)
        else:
          type=w["type"]
          c=str("1")
        obj=Wafer(w["type"]+"-"+c,[i for i in w["processing_times"] ],w["processing_times"])
        wafers.append(obj)
        event_queue.append(obj)


#assigning
while event_queue:
  event=event_queue.popleft()
  all_steps=set()
  allPossibleSteps(event,dag,all_steps)

  #choose machine
  min=float('inf')
  for a in all_steps:
    for j in machines[a]:
      if max(j.end_time,event.ready_time) < min:
        m=j
        step=a
        min=max(j.end_time,event.ready_time)


  #check parameters and allocate
  if not inRange(steps_parameters[step],m.curr_parameters):
    cooldown(m)
  m.end_time=event.processing_time[step]+max(event.ready_time,m.end_time)
  m.processed_wafers+=1
  print(m.m_id,event.w_id,m.end_time,event.ready_time)
  d={}
  d["wafer_id"]=event.w_id
  d["step"]=step
  d["machine"]=m.m_id
  d["start_time"]=m.end_time-event.processing_time[step]
  d["end_time"]=m.end_time

  res.append(d)

  #fluctuations
  if m.processed_wafers%m.n==0:
    for p in m.curr_parameters.keys():
      m.curr_parameters[p]+=m.fluctuation[p]
  event.step_id.remove(step)
  if len(event.step_id)>0:
    event.ready_time=m.end_time
    event_queue.append(event)
  event.completed_step.append(step)


#find time
larger=float('-inf')
for m in machines.values():
  for i in m:
    if i.end_time>larger:
      larger=i.end_time
print(larger)
print(res)
```
